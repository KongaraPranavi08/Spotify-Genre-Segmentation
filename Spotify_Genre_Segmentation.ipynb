```python
features = [
    'danceability',
    'energy',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo'
]

X = df[features]

print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)

df['Cluster'] = kmeans.fit_predict(X_scaled)

df[['track_name', 'playlist_genre', 'Cluster']].head()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Cluster', data=df)

plt.title("Number of Songs in Each Cluster")
plt.xlabel("Cluster")
plt.ylabel("Count")

plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print(X_pca.shape)
```

```python
plt.figure(figsize=(10,7))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=df['Cluster'],
    cmap='viridis',
    s=20
)

plt.title("Spotify Song Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

plt.colorbar(label="Cluster")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    x='Cluster',
    hue='playlist_genre',
    data=df
)

plt.title("Cluster Distribution by Playlist Genre")

plt.show()
```

```python
top_playlists = df['playlist_name'].value_counts().head(10).index

temp = df[df['playlist_name'].isin(top_playlists)]

plt.figure(figsize=(15,6))

sns.countplot(
    x='playlist_name',
    hue='Cluster',
    data=temp
)

plt.xticks(rotation=90)

plt.title("Top Playlist Names by Cluster")

plt.show()
```

```python
song_name = input("Enter a Song Name: ")

song = df[df['track_name'].str.lower() == song_name.lower()]

if song.empty:
    print("❌ Song not found in the dataset.")
else:
    cluster = song.iloc[0]['Cluster']
    print(f"\nCluster: {cluster}")

    recommendations = df[df['Cluster'] == cluster][['track_name', 'track_artist']]

    print("\n🎵 Recommended Songs:")
    print(recommendations.head(10))
```
